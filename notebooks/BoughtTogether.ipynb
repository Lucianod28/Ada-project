```python
def make_pairs(data):
    prods = data["PRODUCT_ID"].unique()
    pa = []
    pb = []
    for i in range(len(prods)):
        for j in range(len(prods)):
            if i != j:
                pa.append(prods[i])
                pb.append(prods[j])
    d = pd.DataFrame()
    d["Pairs 1"] = pd.Series(pa, dtype="int32")
    d["Pairs 2"] = pd.Series(pb, dtype="int32")
    return d
```
